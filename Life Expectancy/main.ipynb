```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("Life Expectancy Data.csv")
df
```

```python
df.describe()
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
# %matplotlib inline

# plt.plot(df.Year,color='orange')
df.groupby('Country')['Life expectancy '].mean().plot(kind='bar' ,figsize=(50,15),fontsize=30,color='blue')
```

```python
from sklearn.preprocessing import LabelEncoder

lb=LabelEncoder()

df['c']=lb.fit_transform(df.Country)
df['s']=lb.fit_transform(df.Status)
df.drop(['Country','Status'],axis='columns',inplace=True)
df
```

```python
X=df.drop(['Life expectancy '],axis='columns')
Y=df['Life expectancy ']

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

ranf=RandomForestRegressor(n_estimators=150)

ranf.fit(x_train,y_train)
ranf.score(x_test,y_test)
```
